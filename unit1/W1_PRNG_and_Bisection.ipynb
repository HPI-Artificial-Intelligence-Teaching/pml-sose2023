```julia
using Plots

function PRNG(X0, n;  a = 7^5, c=0, m=2^31-1)
    X = Vector{Float64}(undef, n)
    X[1] = X0 
    for i in 2:n
        X[i] = mod(a*X[i-1]+c, m)
    end
    return X ./ m
end

X = PRNG(1234, 100000)

histogram(X, bins = 100)
```

```julia
function Bisection(f, a, b, eps)
    f(a)*f(b) < 0 || error("a and b must have opposing signs")
    c = (a + b)/2
    while abs(f(c)) > eps
        if sign(f(c)) == sign(f(a))
            a = c
        else
            b = c
        end
        c = (a+b)/2
    end
    return c
end

f(x) = sin(x)

Bisection(f, -2, 1, 0.0000001)
```
